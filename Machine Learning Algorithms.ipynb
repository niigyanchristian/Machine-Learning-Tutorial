# Machine Learning Algorithms

**What's Next in Machine Learning?**

Now, let's delve into learning algorithms, starting with **Supervised Learning**.

**1. Regression Analysis:**
Regression is a powerful technique for predicting a continuous outcome variable based on one or more predictor variables. It helps us understand the relationship between input variables and the target variable, enabling predictions on unseen data.

Types:
- **a. Linear Regression:** One of the most popular algorithms, showcasing a linear relationship between dependent and independent variables.
    - i. Simple Linear Regression
    - ii. Multiple Linear Regression
    - iii. Multivariate Linear Regression
- **b. Polynomial Regression:** Fits a curve to the data, capturing complex relationships.
- **c. Logistic Regression:** Predicts categorical values.
- **d. Support Vector Regression:** Extends the principles of Support Vector Machines (SVM) to solve regression problems.

**2. Classification Algorithms:**
Classification algorithms categorize new observations based on training data. They're used for binary or multi-class classification tasks.

- **Binary Classifier:** Two possible outcomes (e.g., Yes or No, ON or OFF, -1 or 1, 0 or 1).
- **Multi-class Classifier:** More than two outcomes (e.g., sports, politics, entertainment, education).
- **Multi-label Classification:** So X can potentially belong to several categories.

Types:
1. **Linear Models:**
    - Logistic Regression
    - Support Vector Machine (SVM)\
    
2. **Non-linear Models:**
    - K-nearest Neighbor
    - Kernel SVM
    - Naive Bayes
    - Decision Tree Classification
    - Random Forest Classification

**Evaluation Metrics:**
- **Log Loss or Cross-Entropy Loss:** Measures classification model performance.
- **Confusion Matrix:** Visualizes model performance.
- **AUC-ROC Curve:** Measures classifier's ability to distinguish between classes.

*Moving Forward:*
We'll now dive into implementing these algorithms.
1. **Logistic Regression:** Predicts categorical dependent variables using independent variables.
2. **K-nearest Neighbor (K-NN):** A lazy learner, classifies based on the similarity of stored data.
3. **Support Vector Machine (SVM):** Primarily used for classification problems.
4. **Naive Bayes:** Simple yet effective algorithm for classification tasks.
