## Introduction to Activation Functions

Before we jump into the types of activation functions, let's understand their purpose. Activation functions are mathematical equations that determine the output of a neural network node. They introduce non-linearity into the network, enabling it to learn complex patterns in the data.

### Types of Activation Functions

Now, let's explore some common types of activation functions used in neural networks.


1. **Sigmoid Function:**

The sigmoid function, also known as the logistic function, squashes the input values between 0 and 1. It's useful for binary classification tasks, but it suffers from the vanishing gradient problem, making it less suitable for deep neural networks.



2. **Hyperbolic Tangent (Tanh) Function:**

Similar to the sigmoid function, the tanh function squashes input values between -1 and 1. It overcomes some of the limitations of the sigmoid function and is commonly used in hidden layers of neural networks.


3. **Rectified Linear Unit (ReLU):**

ReLU is one of the most popular activation functions, especially in deep learning. It outputs the input directly if it's positive, and zero otherwise. ReLU helps mitigate the vanishing gradient problem and accelerates convergence during training.


4. **Leaky ReLU:**

Leaky ReLU is a variation of ReLU that allows a small, positive gradient when the input is negative. This helps address the "dying ReLU" problem, where neurons can become inactive during training.



5. **Exponential Linear Unit (ELU):**

ELU is another variation of ReLU that avoids the dying ReLU problem by allowing negative values. It introduces an exponential component for negative inputs, which smoothens the transition.
